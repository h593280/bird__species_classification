```python
import os
import soundfile as sf

# Define the path to the root directory containing the audio files
root_path = "../dataset/audio"

# Define the name of the directory to process
directory_name = "Grønnfink"

# Get the path to the directory to process
directory = os.path.join(root_path, directory_name)

# Get a list of all audio files in the directory
audio_files = [f for f in os.listdir(directory) if f.endswith(".mp3")]

# Loop through each audio file and convert it to WAV
for audio_file in audio_files:
    mp3_path = os.path.join(directory, audio_file)
    wav_path = os.path.splitext(mp3_path)[0] + ".wav"

    print(f"Converting {mp3_path} to {wav_path}")

    # Load the MP3 file using soundfile
    audio_data, sample_rate = sf.read(mp3_path)

    # Export the audio as WAV
    sf.write(wav_path, audio_data, sample_rate)
```
